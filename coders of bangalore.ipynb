```python
def parse_chunk(chunk):
    try:
        chunk = chunk.strip()
        lines = chunk.split('\n')
        
        # Initialize variables
        username = None
        no_of_posts = 0
        no_of_followers = 0
        no_of_following = 0
        name = None
        type_of_page = "Unknown"
        bio = ""
        
        # Process the first line which contains username and possibly other info
        if lines:
            # The first line might contain the full name and other details
            first_line = lines[0].strip()
            if '|' in first_line:
                # Format: "Username | Other info"
                parts = first_line.split('|', 1)
                username = parts[0].strip()
            else:
                username = first_line
        
        # Find posts, followers, and following counts
        for i, line in enumerate(lines):
            line = line.strip()
            
            if "post" in line:
                count_text = line.split("post")[0].strip()
                no_of_posts = convert_to_number(count_text)
            
            elif "follower" in line:
                count_text = line.split("follower")[0].strip()
                no_of_followers = convert_to_number(count_text)
            
            elif "following" in line:
                count_text = line.split("following")[0].strip()
                no_of_following = convert_to_number(count_text)
                
                # After finding following, the next line is likely the page type
                if i + 1 < len(lines):
                    type_of_page = lines[i + 1].strip()
                
                # And everything after that is the bio
                if i + 2 < len(lines):
                    bio = "\n".join(lines[i + 2:])
                
                break  # Stop after finding following as we've processed all counts
        
        # Try to extract name - it's usually the first line if not already parsed as username
        if not name and username:
            name = username
        
        return {
            "username": username,
            "no_of_posts": no_of_posts,
            "no_of_followers": no_of_followers,
            "no_of_following": no_of_following,
            "name": name,
            "type_of_page": type_of_page,
            "bio": bio
        }
    
    except Exception as e:
        print(f"Error parsing chunk: {e}")
        print(f"Problematic chunk: {chunk}")
        return None

def convert_to_number(text):
    """Convert string numbers with K, M, B suffixes to actual numbers"""
    if not text:
        return 0
    
    # Remove commas and spaces
    text = text.replace(",", "").strip()
    
    # Handle cases like "1.9M"
    if text[-1].upper() in ['K', 'M', 'B']:
        multiplier = {
            'K': 1000,
            'M': 1000000,
            'B': 1000000000
        }[text[-1].upper()]
        
        # Extract the number part
        number = float(text[:-1].strip())
        return int(number * multiplier)
    
    # Handle regular numbers
    return int(float(text))
```
